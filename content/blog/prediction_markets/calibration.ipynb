```python
# Correcting the data points, adding the final point (100, 100), and merging all adjustments in a single code block

import matplotlib.pyplot as plt

# Correct data points with (1, 1) at the start and (100, 100) at the end
pairs = [
    (99, 99), (97, 97), (95, 91), (90, 82), (80, 74), (70, 64),
    (60, 55), (50, 46), (40, 36), (30, 26), (20, 18), (10, 9), (1, 1),
]
x_values, y_values = zip(*pairs)

# Plotting with all specified requirements
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='red', label='Kalibrációs görbe')
plt.plot([0, 100], [0, 100], linestyle='--', color='black', linewidth=1, label='Tökéletes kalibráció')

# Hungarian labels and title
plt.xlabel('Előrejelzett Valószínűség (%)')
plt.ylabel('Bekövetkezettek Aránya (%)')
plt.title('Manifold Markets Kalibrációja')
plt.grid(True)

# Adding annotations slightly further to the right of each point in black
for i in range(len(x_values)):
    plt.text(x_values[i] + 2, y_values[i], f"({x_values[i]}, {y_values[i]})", fontsize=8, ha='left', color='black')

# Adding legend
plt.legend(loc='upper left')

# Save the corrected plot as SVG
plt.savefig('C:/Users/takat/OneDrive/Documents/Programming/Blog/eti-blog/static/prediction_markets/manifold.svg', format='svg')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
file_path = 'polymarket.csv'
data = pd.read_csv(file_path)

data['Date (UTC)'] = pd.to_datetime(data['Date (UTC)'], format='%m-%d-%Y %H:%M')

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(data['Date (UTC)'], data['Price'], linestyle='-', label='Árfolyam')

# Add a vertical line for Biden's withdrawal
biden_withdrawal_date = pd.to_datetime('2024-07-21')
plt.axvline(x=biden_withdrawal_date, color='red', linestyle='--', label='Biden kiszáll a versenyből')

# Hungarian labels and title
plt.title('A "Biden drops out of presidential race?" fogadás árfolyama')
plt.xlabel('Dátum')
plt.ylabel('Árfolyam')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.tight_layout()

# Save the plot as SVG
plt.savefig('C:/Users/takat/OneDrive/Documents/Programming/Blog/eti-blog/static/prediction_markets/polymarket_with_biden.svg', format='svg')

plt.show()
```

```python
import yfinance as yf
import pandas as pd
from datetime import timedelta

# Load your CSV
df = pd.read_csv("data/spcegek.csv")

# Function to download stock data 2 months from "Date added"
def download_stock_data(symbol, date_added):
    date_added = pd.to_datetime(date_added)
    start_date = date_added - pd.DateOffset(months=12)
    end_date = date_added + pd.DateOffset(months=2)
    
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data

# Create an empty dictionary to hold the data
stock_data_dict = {}

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    symbol = row['Symbol']
    date_added = row['Date added']
    
    try:
        stock_data = download_stock_data(symbol, date_added)
        stock_data_dict[symbol] = stock_data
    except Exception as e:
        print(f"Could not download data for {symbol}: {e}")

# You can save the data for each stock to CSV if needed
for symbol, data in stock_data_dict.items():
    data.to_csv(f'data/{symbol}_stock_data.csv')
```

```python
import pandas as pd
import numpy as np

# Create an empty DataFrame
df = pd.DataFrame()

# Find the maximum length of the 'Close' price arrays
max_len = max(len(data['Close']) for data in stock_data_dict.values())

# Iterate through the stock data dictionary
for symbol, data in stock_data_dict.items():
    close_prices = np.array(data['Close'])
    
    # Pad the shorter arrays with NaNs to match the max_len
    padded_prices = np.pad(close_prices, (0, max_len - len(close_prices)), constant_values=np.nan)
    
    # Add the padded array to the DataFrame
    df[symbol] = padded_prices

# Reset the index to be just numeric without keeping the old one
df.reset_index(drop=True, inplace=True)

# Display the result
print(df)
```
